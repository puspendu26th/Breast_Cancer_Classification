# Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

# Data Collection and Processing

```python
# loading the data from sklearn
breastcancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breastcancer_dataset)
```

```python
data_frame = pd.DataFrame(breastcancer_dataset.data, columns = breastcancer_dataset.feature_names)
data_frame
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breastcancer_dataset.target
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of target Varibale
data_frame['label'].value_counts()
```

1 for Benign & 0 for Malignant

```python
data_frame.groupby('label').mean()
```

# Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

# Splitting the data into Training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

# Standardize the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

# Importing Tensorflow

```python
# importing tensorflow and Keras
!pip install tensorflow-gpu
import tensorflow as tf 
tf.random.set_seed(3)
from tensorflow import keras
from tensorflow.keras.optimizers import Adam
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([keras.layers.Flatten(input_shape=(30,)), keras.layers.Dense(20, activation='relu'),keras.layers.Dense(2, activation='sigmoid')])
                          
                         
                          
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
              
             
```

```python
# training the Meural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=100)
```

# Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

# Built the predictive system

```python
input_data = (10.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)


input_data_as_numpy_array = np.asarray(input_data)


input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)


input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 1):
  print('The Tumor is Benign')

else:
  print('The Tumor is Malignant')
```
